```
import pandas as pd
```

```
# 데이터 불러와서 필요한 변수 뽑기
```

```
df1_1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/seouldata/서울시 우리마을가게 상권분석서비스(상권-점포).csv", encoding='CP949')
df1_1
```

```
df1_0 = df1_1[['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명','점포_수','개업_점포_수','폐업_점포_수']]
df1_0
```

```
df1_2 = df1_0.groupby(['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명']).sum()
df1_2
```

```
df1_2 = df1_2.reset_index(['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'])
df1_2
```

```
df2_1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/seouldata/서울시 우리마을가게 상권분석서비스(상권-추정매출).csv", encoding='CP949')
df2_1
```

```
df2_0 = df2_1[['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명','당월_매출_금액','당월_매출_건수']]
df2_0
```

```
df2_2 = df2_0.groupby(['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명']).sum()
df2_2
```

```
df2_2 = df2_2.reset_index(['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'])
df2_2
```

```
df3_1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/seouldata/서울시 우리마을가게 상권분석서비스(상권-추정유동인구).csv", encoding='CP949')
df3_1.head()
```

```
is_2020 = df3_1['기준_년_코드'] == 2020
df3 = df3_1[is_2020]
df3
```

```
df3_0 = df3[['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명','총_유동인구_수']]
df3_0
```

```
# 중간 과정: 변수 합치기
```

```
new0_df = pd.merge(df3_0,df2_2, on=['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'], how='outer')
new0_df
```

```
new00_df = pd.merge(new0_df,df1_2,on=['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'], how='outer')
```

```
df4_1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/seouldata/서울시 우리마을가게 상권분석서비스(상권-집객시설).csv", encoding='CP949')
df4_1
```

```
df5_1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/seouldata/서울시 우리마을가게 상권분석서비스(상권-직장인구).csv", encoding='CP949')
df5_1
```

```
df5_0 = df5[['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명','총_직장_인구_수']]
df5_0
```

```
df6_1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/seouldata/서울시 우리마을가게 상권분석서비스(상권-아파트).csv", encoding='CP949')
df6_1
```

```
is_2020 = df6_1['기준_년_코드'] == 2020
df6 = df6_1[is_2020]
df6
```

```
df7_1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/seouldata/서울시 우리마을가게 상권분석서비스(상권-상권변화지표).csv", encoding='CP949')
df7_1
```

```
df7_0 = df7[['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명','상권_변화_지표','상권_변화_지표_명','운영_영업_개월_평균','폐업_영업_개월_평균']]
df7_0
```

```
# 뽑아낸 변수들을 하나의 데이터로 합치기
```

```
new1_df = pd.merge(df4_0,df5_0, on=['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'], how='outer')
new1_df
```

```
new11_df = pd.merge(df6_0,df7_0, on=['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'], how='outer')
new11_df
```

```
new2_df = pd.merge(new1_df,new11_df, on=['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'], how='outer')
new2_df
```

```
new_df = pd.merge(new00_df,new2_df, on=['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'], how='outer')
new_df
```

```
new_df = new_df.set_index(['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'])
new_df
```
