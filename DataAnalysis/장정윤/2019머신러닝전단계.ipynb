```
# train과 test를 7:3으로 분할
```

```
test = da19.sample(frac=0.3,replace=False,random_state=1)
train = da19.sample(frac=0.7,replace=False,random_state=7)
```

```
train = train.drop(['기준_년_코드','기준_분기_코드','상권_코드','상권_코드_명'], axis=1)
train
```

```
# minmaxscaler을 이용해 모든 값을 0과 1사이로 변환
```

```
scaler = MinMaxScaler()
train[:] = scaler.fit_transform(train[:])
```
