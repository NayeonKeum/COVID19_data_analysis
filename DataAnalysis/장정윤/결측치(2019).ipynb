```
import pandas as pd
```

```
data1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/seouldata/data2019.csv", encoding='utf-8-sig')
data1
```

```
# 결측값을 0으로 대체
```

```
data1['집객시설_수'] = data1['집객시설_수'].fillna(0)
data1
```

```
data1['총_직장_인구_수'] = data1['총_직장_인구_수'].fillna(0)
data1
```

```
data1['아파트_단지_수'] = data1['아파트_단지_수'].fillna(0)
data1
```
