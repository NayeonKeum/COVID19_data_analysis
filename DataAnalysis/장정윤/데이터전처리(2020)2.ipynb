```
import pandas as pd
```

```
# 상권변화지표를 1,0으로 표현, 변수 제거
```

```
dd = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/seouldata/data2020_1.csv")
dd
```

```
dd['상권_변화_지표'].replace("HH", 1, inplace=True)
dd
```

```
dd['상권_변화_지표'].replace("LL", 1, inplace=True)
dd
```

```
dd['상권_변화_지표'].replace("LH", 1, inplace=True)
dd
```

```
dd['상권_변화_지표'].replace("HL", 0, inplace=True)
dd
```

```
dd = dd.drop(["상권_변화_지표_명","운영_영업_개월_평균","폐업_영업_개월_평균"], axis=1)
dd
```

```
dd = dd.set_index(["기준_년_코드","기준_분기_코드","상권_코드","상권_코드_명"]) 
dd
```

```
dd.to_csv("/content/drive/My Drive/Colab Notebooks/seouldata/data2020_3.csv")
```
